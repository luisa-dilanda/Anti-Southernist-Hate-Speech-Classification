For each keyword that we selected, we create here a list that'll comprehend all the articles that we can retrieve using that keyword as a query on the websites that we chose in the beginning. We use the sleep function so as not to be blocked by those web pages

We create a function for the creation of the lists and another one to save the list in a text file. This was necessary, because the crawling of the articles took a bit of time, so the operation of finding the links + getting the corresponding articles' text + labeling the texts couldn't be done in one go. 
By saving the links in a text file, we were able to then retrieve the texts from the corresponding links without having to go by the process of retrieving the links all over again. 

```python
def listoflinks(listofwebsites):
    l=[]
    for i in range(len(listofwebsites)):
        sleep(randint(1, 5))
        giornale = newspaper.build(listofwebsites[i])
        for article in giornale.articles:
            l.append(article.url)
    return l
```

```python
def savelist(lista,nomefile):
#SALVA LISTA SU FILE DI TESTO
    with open(nomefile, "w") as f:
        for s in lista:
            f.write(str(s) +"\n")
```

# Let's create a function that, given a list of links, creates a dataframe with the links, keyword, text and label
(the label will be assigned later)

```python
from newspaper import Article
```

```python
def build_df(lista,keyword):
    #build a dataframe for each list
    df = pd.DataFrame(lista)
    #chiama "link" la prima colonna
    df.rename(columns={0: "link"}, inplace=True)
    #aggiungi colonne per parola chiave cercata (keyword), testo dell'articolo (text) ed etichetta binaria antimeridionalismo (label)
    cols = ['keyword', 'text', 'label']
    df = pd.concat([df, pd.DataFrame(columns=cols)])
    df = df.assign(keyword=keyword)
    for i in range(len(df)):
        url = df['link'][i]
        a = Article(url, language='it')
        a.download()
        try:
            a.parse()
            df['text'][i]=a.text
        except:
            pass
    return df
```

```python
def indicilungh(lungh,df):
    n=len(df)
    ind = [0]*n
    for i in range(len(df)):
        vl=[0]
        text=df['text'][i] #assegno il testo dell'articolo i-mo a text
        frasi = sent_tokenize(text) #tokenizzo in frasi il testo di text
        paroleprimafrase=custom_tokenizer.tokenize(frasi[0]) #tokenizzo in parole la prima frase del testo 
        z=len(paroleprimafrase) #e ne segno la lunghezza
        vl.append(z)
        for j in range(1, len(frasi)):
            paroleperfrase=custom_tokenizer.tokenize(frasi[j])
            x=len(paroleperfrase)
            y=vl[-1]+x
            vl.append(y) #quindi a questo punto vl ha le lunghezze cumulative delle frasi dell'articolo i-mo
        ind[i]=list(x >= lungh for x in vl).index(True) #metto nell'i-mo elemento del vett ind l'indice in cui supero le 512 parole
    return ind
```

```python
def separatesti(ind,df,newdf):
    warnings.filterwarnings('ignore')
    for i in range(len(df)):
        text=df['text'][i] #assegno il testo a text
        frasi=sent_tokenize(text) #tokenizzo in frasi il testo di text
        n=df['label'][i]
        if n:
            text1=' '.join(frasi[:ind[i]-1])
            text2=' '.join(frasi[ind[i]-1:])
        else:
            text1=' '.join(frasi[:ind[i]-1])
            text2=' '.join(frasi[:ind[i]-1])
        words1=custom_tokenizer.tokenize(text1)
        words2=custom_tokenizer.tokenize(text2)
        newdf['text'][2*i]=text1
        newdf['text'][2*i+1]=text2
        newdf['textlen'][2*i]=len(words1)
        newdf['textlen'][2*i+1]=len(words2)        
    newdf = newdf.drop_duplicates(subset='text', keep="first", ignore_index=True)
    return newdf
```

```python
def remove_text_duplicates(df):
    df = df.drop_duplicates(subset='text', keep="first") #rimuovi i testi duplicati
    df = df[df['text'].notna()] #remove NaNs
    df = df.reset_index(drop=True) #reset index
    return df
```

```python

def calcola_flag(df, listofwords):
    warnings.warn("Some warning!")
    for word in listofwords:
        for i in range(len(df)):
            if word in str(df['text'][i]).lower():
                df['flag'][i]=1
    tot = df['flag'].sum()
    
    return df, tot
```
